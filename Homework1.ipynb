```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define a list and array
myList = [1,2,3,4,5,6,7]

myArray = np.array([1,2,3,4,5,6,7])
```

```python
#square a list function
def square_list(input):
    return [i**2 for i in input]
```

```python
#print a squared list
squaredList = square_list(myList)
print(squaredList)
```

```python
#make a matplotlib plot

plt.figure()
plt.plot(myList, label='Original Data')
plt.plot(squaredList, label='Squared Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Plot')
plt.legend()
plt.grid(True)

plt.show()
```
